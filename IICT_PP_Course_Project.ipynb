```python
!pip install bokeh

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook

output_notebook()

```

```python
from google.colab import files

file_name = "/content/drive/MyDrive/Python_CSV/Pop_Division.csv"

df = pd.read_csv(file_name)

df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, y='Population (in millions)', x='Area (in sq. km)', marker='o')
plt.title('Line Graph: Population (in millions) vs Area (in sq. km)')
plt.ylabel('Population (in millions)')
plt.xlabel('Area (in sq. km)')
plt.show()
```

```python
plt.bar("Area (in sq. km)","Population (in millions)", data=df)
plt.title('Bar Graph: Population (in millions) vs Area (in sq. km)')
plt.grid(color = "green", linestyle = "--", linewidth = "0.5")
```

```python
category_counts = df['Area (in sq. km)'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Graph: Category Distribution')
plt.show()
```
