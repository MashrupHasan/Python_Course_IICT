```python
!pip install bokeh

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook

output_notebook()

```

```python
from google.colab import files

file_name = "/content/drive/MyDrive/Python_CSV/environmental_data.csv"

df = pd.read_csv(file_name)

df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, y='Temperature', x='Humidity', marker='o')
plt.title('Line Graph: Temperature vs Humidity')
plt.ylabel('Temperature')
plt.xlabel('Humidity')
plt.show()
```

```python
plt.bar("Temperature","Pressure", data=df)
plt.title('Bar Graph: Pressure vs Temperature')
plt.grid(color = "green", linestyle = "--", linewidth = "0.5")
```

```python
category_counts = df['Humidity'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Graph: Humidity')
plt.show()
```
